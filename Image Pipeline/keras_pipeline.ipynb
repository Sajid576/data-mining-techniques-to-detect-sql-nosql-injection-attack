```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
from PIL import Image
from tensorflow.keras.layers import Dropout, Flatten, Dense
from tensorflow.keras.models import Sequential
```

```python
def save_bottlebeck_features():
    name = ['train', 'val', 'test']
    tag = range(1, 11)
    data = {'train':[], 'val':[], 'test':[]}
    labels = {'train': [], 'val': [], 'test': []}
    data_path="/content/drive/MyDrive/My Colab Notebook/Image Pipeline/dataset/"

    for i in name:
        for j in tag:
            path = data_path+str(i)+"/"+str(j)+"/"
            for _, file in enumerate(os.listdir(path)):
                print(path)
                image = Image.open(path + file).convert('L')
                image = image.resize((28, 28))
                image = np.array(image)
                #image = np.true_divide(image, 255)
                data[i].append(image)
                labels[i].append(j)

    labels['train'], labels['val'], labels['test'] = np.array(labels['train']), np.array(labels['val']), np.array(labels['test'])
    
    np.savez('mnistlikedataset.npz', train=data['train'], val=data['val'], test=data['test'])

    path = 'mnistlikedataset.npz'
    with np.load(path) as data:
        return data['train'], data['val'], data['test'],labels['train'], labels['val'], labels['test']
```
